```python
data['Year'] = pd.DatetimeIndex(data['event_time']).year
data['Month'] = pd.DatetimeIndex(data['event_time']).month
data['Day'] = pd.DatetimeIndex(data['event_time']).day
data['Date'] = pd.DatetimeIndex(data['event_time']).date
data['Date'] = pd.to_datetime(data['Date'])
data['Day_Name'] = pd.DatetimeIndex(data['event_time']).day_name()
data['YYYYMM'] = data['Date'].dt.strftime('%Y%m')
```

```python
data.groupby(data['event_type']) \
    .size() \
    .reset_index(name='counts') \
    .sort_values(by=['counts'], ascending=False)
```

```python
data[data['event_type'] == 'ReadingOwnedBook'] \
    .groupby(data['account_id']) \
    .size().reset_index(name='counts') \
    .sort_values(by='counts', ascending=False)
```

```python
data[data['event_type'] == 'ReadingOwnedBook'] \
    .groupby(['product_id', 'account_id']) \
    .size().reset_index(name='counts') \
    .sort_values(by='product_id', ascending=False)
```

```python
data[data['event_type'] == 'ReadingOwnedBook'] \
    .groupby(['YYYYMM']) \
    .size().reset_index(name='counts') \
    .sort_values(by='YYYYMM', ascending=True)
```

```python
data[data['event_type'] == 'ReadingOwnedBook'] \
    .groupby(['YYYYMM']) \
    .nunique()
```

```python
data[data['event_type'] == 'ReadingOwnedBook'] \
    .groupby(['YYYYMM']) \
    .agg({'Day': pd.Series.nunique, 'account_id': pd.Series.count}) \
    .rename(columns={'Day':'days', 'account_id':'events'})
```

```python
event_type_summary = data \
    .groupby(['event_type']) \
    .agg({'YYYYMM': pd.Series.nunique, 'Date': pd.Series.nunique, 'product_id': pd.Series.count, 'account_id': pd.Series.nunique}) \
    .rename(columns={'YYYYMM':'months', 'Date':'days', 'product_id':'events', 'account_id': 'customers'}) \
    .sort_values(by=['event_type'])
```

```python
event_type_summary['events_per_account_per_month'] = (event_type_summary['events'] / event_type_summary['customers'])/(event_type_summary['days']/28.0)
```
